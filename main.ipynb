1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.</br>
	•	Qualitative Data: Descriptive data that categorizes or labels (e.g., colors, gender).</br>
	•	Nominal Scale: Data with categories having no order (e.g., colors: red, blue, green).</br>
	•	Ordinal Scale: Data with categories that can be ordered but no meaningful difference (e.g., movie ratings: 1 star, 2 stars, etc.).</br>
	•	Quantitative Data: Numerical data representing quantities (e.g., height, weight).</br>
	•	Interval Scale: Data with ordered values and meaningful differences, but no true zero (e.g., temperature in Celsius).</br>
	•	Ratio Scale: Data with a true zero and meaningful ratios (e.g., weight, height).</br>
</br>
2. What are the measures of central tendency, and when should you use each?</br>
	•	Mean: Average of the data, used when there are no extreme outliers.</br>
	•	Median: Middle value, used when data has outliers or skewness.</br>
	•	Mode: Most frequently occurring value, used for categorical data or finding the most common value in a dataset.</br>
</br>
3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?</br>
	•	Dispersion: Refers to the spread of data values around a central value.</br>
	•	Variance: Measures the average squared deviation from the mean.</br>
	•	Standard Deviation: Square root of variance, represents average deviation from the mean in original units.</br>
</br>
4. What is a box plot, and what can it tell you about the distribution of data?</br>
	•	A box plot visualizes the data’s spread and highlights the median, quartiles, and potential outliers.</br>
	•	It shows:</br>
	•	Central value (median),</br>
	•	Spread (interquartile range, or IQR),</br>
	•	Extreme values (outliers).</br>
</br>
5. Discuss the role of random sampling in making inferences about populations.</br>
	•	Random sampling ensures every member of a population has an equal chance of selection.</br>
	•	It reduces bias and allows inferences about the population with known probabilities, enabling statistical analysis and generalizations.</br>
</br>
6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?</br>
	•	Skewness: Measures asymmetry in a dataset.</br>
	•	Positive Skew: Tail on the right; mean > median.</br>
	•	Negative Skew: Tail on the left; mean < median.</br>
	•	Skewness affects central tendency: in skewed data, the median is preferred over the mean.</br>
</br>
7. What is the interquartile range (IQR), and how is it used to detect outliers?</br>
	•	IQR: Difference between the 75th percentile (Q3) and 25th percentile (Q1).</br>
	•	Outliers: Data points are outside[Q1-1.5*IQR,Q3+1.5*IQR].</br>
</br>

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).</br>
	•	Properties:</br>
	•	Symmetrical bell-shaped curve.</br>
	•	Mean = Median = Mode.</br>
	•	Fully defined by mean and standard deviation.</br>
	•	Empirical Rule:</br>
	•	68% of data within 1 SD of mean,</br>
	•	95% within 2 SD,</br>
	•	99.7% within 3 SD.</br>
